### Задание 1:
Обработайте json файл. Он содержит имена пользователей и количество сделанных ими комментариев. Соберите словарь, ключами в котором будут имена пользователей, а значениями - количество сделанных комментариев. Посчитайте общую сумму комментариев.

```python
import json
import re
file = open('comments_326022.json', 'r')
d = json.load(file)
#sum(d.get('count'))
d2 = d['comments']
count = 0
for i in d2:
    print(i['name'] + " " + str(i['count']))
    count += i['count']
print(count)
```
